```python
import pandas as pd

# 데이터 불러오기
df1A = pd.read_csv('D:\\데이터분석 프로젝트\\databass\\1A.csv')
df2A = pd.read_csv('D:\\데이터분석 프로젝트\\databass\\2A.csv')
df3A = pd.read_csv('D:\\데이터분석 프로젝트\\databass\\3A.csv')
df4A = pd.read_csv('D:\\데이터분석 프로젝트\\databass\\4A.csv')
df5A = pd.read_csv('D:\\데이터분석 프로젝트\\databass\\5A.csv')
df6A = pd.read_csv('D:\\데이터분석 프로젝트\\databass\\6A.csv')
df7A = pd.read_csv('D:\\데이터분석 프로젝트\\databass\\7A.csv')
df8A = pd.read_csv('D:\\데이터분석 프로젝트\\databass\\8A.csv')
df9A = pd.read_csv('D:\\데이터분석 프로젝트\\databass\\9A.csv')

# 콘캣으로 데이터 묶기
row_concat = pd.concat([df1A, df2A, df3A, df4A, df5A, df6A, df7A, df8A, df9A])
db = pd.DataFrame(row_concat)

# 행 이름 재지정
db.rename(columns={
  "STD_YM": "month",
	"BLCK_SP_CD": "classcode" ,
	"CTPV_CD": "citycode" ,
	"CTPV_NM": "city",
	"CARD_HLDR_CT": "user" ,
	"AVG_CARD_CNT": "avg_card" ,
	"AVG_CARD_LMT_AMT": "lmt" ,
	"CARD_USE_AMT": "use" ,
	"CRD_SLE_USE_AMT": "sle" ,
	"LSP_USE_AMT": "lsp" ,
	"INSTL_USE_AMT": "instl" ,
	"CASH_SVC_SUE_AMT": "cash" ,
  "OVSEA_CSMP_AMT": "global"}, inplace=True)
db
  
```

<h1> 서울 </h1>

```python
seoul_a=db[db['city'].isin(['서울특별시'])]
seoul_a.set_index(['city', 'month'], inplace=True)
seoul=seoul_a.loc[:, ['user',	'avg_card',	'lmt',	'use',	'sle',	'lsp',	'instl',	'cash',	'global']]
seoul
```

```python
html = seoul.to_html()
print(html)
```
